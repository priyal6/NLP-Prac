```python
df = pd.read_csv('Output.csv')
df.head()
```

```python
df['review'] = df['review'].apply(lambda txt: txt.lower())
```

```python
stop_words = nltk.corpus.stopwords.words('english')
```

```python
df['review'] = df['review'].apply(lambda txt: ' '.join([word for word in txt.split() if word not in stop_words]))
```

```python
#tokenization
df['review'] = df['review'].apply(lambda txt: sent_tokenize(txt))
```

```python
df['review'] = df['review'].apply(lambda txt: ' '.join(txt))
```

```python
sia = SentimentIntensityAnalyzer()
df['vader_score'] = df['review'].apply(lambda txt: sia.polarity_scores(txt))
```

```python
df.head(3)
```

```python
df['negative'] = df['vader_score'].apply(lambda txt: txt['neg'])
df['neutral'] = df['vader_score'].apply(lambda txt: txt['neu'])
df['positive'] = df['vader_score'].apply(lambda txt: txt['pos'])
df['compound'] = df['vader_score'].apply(lambda txt: txt['compound'])
```

```python
df.head()
```

```python
def polarity_score(compound):
  if compound > 0:
    return 'positive'
  elif compound < 0:
    return 'negative'
  else:
    'neutral'
```

```python
df['sentiment'] = df['compound'].apply(lambda val: polarity_score(val))
df.head(3)
```
